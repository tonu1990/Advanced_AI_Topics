### **Auto Encoders**

An autoencoder is a type of artificial neural network that learns to represent **data in a compressed form and then reconstructs it** as closely as possible to the original input.

* Autoencoders consists of three components:

    * **Encoder**: This compresses the input into a compact representation and capture the most relevant features.

    * **Latent Space** (Bottleneck) : The compressed, encoded representation of the input.

    * **Decoder**: It reconstructs the input data from this compressed form to make it as similar as possible to the original input.

This part contains the most important features of the input data.
* Autoencoders aim to minimize reconstruction error which is the difference between the input and the reconstructed output
* It is an **unsupervised learning technique** â€” you don't need labeled data to train it.
* One main variation of AutoEncoder is in Computer Vision - **Convolutional Autoencoder** - This type of autoencoder uses convolutional layers instead of fully connected layers.
* Applications :
  * Dimensionality Reduction (like PCA but nonlinear)
  * Denoising images or signals
  * Anomaly Detection (e.g., in manufacturing or security logs)
  * Data Compression
  * Generating new data (with some types like Variational Autoencoders)
